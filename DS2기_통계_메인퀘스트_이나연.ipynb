## 미션2) 'Country' 변수를 바탕으로 가장 빈도가 높은 두 개의 국가를 선택하고, 각 국가의 'TotalPrice' 평균에 차이가 있는지 검증하세요. 독립표본 t-검정을 사용하여 가설을 설정하고 검정하며, 1종 오류와 2종 오류에 대해 설명하세요.


* 미션 진행 방법
    <br>1) 데이터 로드 후 'TotalPrice'와 'Country' 컬럼을 찾아주세요.
    <br>2) 데이터셋에서 주문이 가장 많은 상위 두 국가를 찾아주세요.
    <br>3) 통계적 검증 수행하기
    <br>4) 각 국가의 'TotalPrice' 평균을 계산하세요.
    <br>5) 독립표본 t-검정을 사용하여 두 국가 간 'TotalPrice' 평균의 차이가 통계적으로 유의미한지 검증하세요. 이 때, scipy.stats 라이브러리의 ttest_ind 함수를 사용할 수 있습니다. equal_var=False 파라미터를 설정하여 두 집단의 분산이 다르다고 가정하세요.
    <br>6) t-통계량과 p-값을 해석하여 결과를 설명하세요. 1종 오류와 2종 오류의 개념을 설명하고, 해당 데이터셋에 적용하여 설명해 주세요.

````python
# df의 totalprice랑 country 컬럼만 보기
ct = df[['TotalPrice', 'Country']]

# 각 국가의 totalprice의 평균 계산
country_means = ct.groupby('Country')['TotalPrice'].mean()
print(country_means)
````

````python
# top2 국가 찾기
country_counts = df['Country'].value_counts()
print('top2 country:\n', country_counts.head(2))
````

````python
# 통계적 검정하기 - 독립표본 t 검정
from scipy import stats

# top2 국가 변수 설정
top_two_countries = country_counts.head(2).index
c1, c2 = top_two_countries[0], top_two_countries[1] #c1 = country1, c2 = country2

# 두 국가의 주문 totalprice
order_c1 = df[df["Country"] == c1]["TotalPrice"]
order_c2 = df[df["Country"] == c2]["TotalPrice"]

# 두 국가의 주문 totalprice 평균, 표준편차
c1_mean = order_c1.mean()
c1_std = order_c1.std()
c2_mean = order_c2.mean()
c2_std = order_c2.std()

print(f'{c1}\nmean: {c1_mean}, std: {c1_std}\n')
print(f'{c2}\nmean: {c2_mean}, std: {c2_std}\n')

# 독립 표본 t-검정
t_stat, p_value = stats.ttest_ind(order_c1, order_c2, equal_var=False)
print(f"T-statistic: {t_stat},\nP-value: {p_value}")

# 결과
alpha = 0.05
if p_value < alpha:
    print("두 국가 간의 주문량 평균에 통계적으로 유의미한 차이가 있다.")
else:
    print("두 국가 간의 주문량 평균에 통계적으로 유의미한 차이가 없다.")
````

## 1종 오류 (Type I error)
: 귀무가설이 참일때, 귀무가설을 기각하는 오류

즉, 귀무가설이 실제로 참이라 채택해야하는데, 표본의 오차때문에 채택하지 않는 오류이다.

통계 가설 검정 시 1종 오류를 1% 또는 5% 이내로 보통 설정한다. (유의수준, significance level)

## 2종 오류 (Type II error)
: 귀무가설이 거짓일때, 귀무가설을 채택하는 오류

즉, 귀무가설이 거짓이라 채택하지 말아야하는데 표본의 오차때문에 채택하는 오류이다. 치명적 오류라고 한다.

검정력(statistical power)은 1-2종 오류 발생 확률을 뺀 것이다.

\
## 해당 데이터셋에 적용

P-value가 매우 작게 나왔기에 1종 오류가 발생할 가능성은 낮기에 top2 국가인 United Kingdom과 Germany 간의 주문량 평균에 통계적으로 유의미한 차이가 있다고 판단 가능하다.

P-value가 보통 유의수준으로 생각하는 0.05보다 높았다면, 유의하지 않았다고 판단했을 가능성이 높다. 이 경우에는 2종 오류를 생각해보면 된다.

## 미션4) 심슨의 역설을 설명하기 위한 구체적인 분석을 다음의 순서로 진행해 주세요.

심슨의 역설(Simpson's Paradox)은 통계 데이터를 해석할 때 발생할 수 있는 현상으로, 일부 하위 그룹에서 관찰된 추세가 전체 데이터를 합쳤을 때 반전될 수 있음을 나타냅니다. 이 현상은 복잡한 데이터에서 하위 그룹 간에 상이한 분포나 크기 때문에 발생할 수 있습니다.

심슨의 역설을 설명하기 위해서는, 데이터 내에서 서로 상반된 추세를 보이는 하위 그룹을 찾거나 생성해야 합니다. 예를 들어, 특정 상품의 판매량이나 단가가 국가별로 다른 추세를 보이는지, 혹은 특정 기간 동안의 판매 추세가 전체 기간 동안의 추세와 다른지 등을 분석할 수 있습니다.

데이터를 제시한 특정 국가로 하위 그룹으로 나눕니다.
각 하위 그룹에서 총 판매금액의 추세를 분석합니다.
전체 데이터셋에 대해 동일한 지표를 분석합니다.
하위 그룹의 추세와 전체 데이터의 추세가 서로 다른 방향성을 보일 경우, 이를 통해 심슨의 역설을 설명합니다.
이러한 분석을 위해, 데이터 내에서 적절한 변수와 하위 그룹을 선정하는 것이 중요합니다. 현재 제공된 데이터만으로는 심슨의 역설을 바로 설명할 수 있는 명확한 예시를 찾기 어려울 수 있지만, 데이터를 깊이 분석하면 관련 현상을 발견할 수도 있습니다.
이를 이용해 비즈니스에 어떻게 활용할 수 있을지도 설명해 주세요.

````python
# 하위 그룹
country_group = ["Japan", "Hong Kong", "Singapore", "Lebanon", "Canada", "USA"]

filter_country = df[df['Country'].isin(country_group)]
````

````python
filter_country
````

````python
 # 하위 그룹에서 총 판매금액의 추세를 분석
country_means = filter_country.groupby('Country')['TotalPrice'].mean()

# 전체 데이터셋에 대해 동일한 지표를 분석
all_country_mean = df.groupby('Country')['TotalPrice'].mean()

# 출력
print('lower country total price sums:\n ', country_means)
print(f'\nall country total price sum:\n {all_country_mean}')
````

````python
# 하위 그룹의 추세와 전체 데이터의 추세가 서로 다른 방향성을 보일 경우, 이를 통해 심슨의 역설을 설명..
# 서로 다른 방향성이라는게 어떻게 표현하면 좋을 지 모르겠어서 t-검정을 사용해 그룹 간의 차이를 검정해봄..

for country in country_group:
    lower_country_tp = filter_country[filter_country['Country'] == country]['TotalPrice']
    all_country_tp = df['TotalPrice']
    t_statistic, p_value = stats.ttest_ind(lower_country_tp, all_country_tp, equal_var=False)

    print(f"Country: {country}")
    print(f"T-statistic: {t_statistic}")
    print(f"P-value: {p_value}\n")
````

````python
def get_remove_outliers(df, column):
    q1 = df[column].quantile(0.25)
    q3 = df[column].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    df_no_outlier = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return df_no_outlier
````

````python
# unitprice 이상치 제거 후 회귀 분석 다시 진행

# 사분위수 계산
q1 = df['UnitPrice'].quantile(0.25)
q3 = df['UnitPrice'].quantile(0.75)

# 사분위수 범위 계산
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

# 이상치 제거
df_no_outlier = df[(df['UnitPrice'] >= lower_bound) & (df['UnitPrice'] <= upper_bound)]
````

````python
df_no_outlier = get_remove_outliers(df, 'TotalPrice')
````
